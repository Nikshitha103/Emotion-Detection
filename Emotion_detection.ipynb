```python
import pandas as pd
```

```python
large_text=pd.read_csv('https://raw.githubusercontent.com/Jcharis/end2end-nlp-project/main/data/emotion_dataset_2.csv')
```

```python
large_text['Clean_Text'][1:10].apply(emotion)
```

```python
def get_emotion_label(text):
  return (emotion (text)[0]['label'])
```

```python
large_text['Clean_Text'][1:10].apply (get_emotion_label)
```
